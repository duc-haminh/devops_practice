```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

data = []
path = "/content/drive/MyDrive/1. Academic/1. University/2. Master/4. Bách Khoa - Data Science/1. Courses/Semester 4/Code/Data/Y Tế/"
filename = "data_thuduchospital"
filepath = path + filename

start_time = time.time()  # Start measuring time

for url in urls:
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all elements with class 'qa__item-user-question' and 'qa__item-admin-answer'
    content_ques_list = soup.find_all('div', class_='qa__item-user-question')
    content_ans_list = soup.find_all('div', class_='qa__item-admin-answer')

    if content_ques_list and content_ans_list:
        for content_ques, content_ans in zip(content_ques_list, content_ans_list):
            excluded_elements = content_ques.find_all('p')
            content_ques_text = ''
            for p in excluded_elements:
                lines = p.get_text(strip=True).split('\n')
                lines = [line for line in lines if line.strip()]
                if lines:
                    content_ques_text += ' '.join(lines) + ' '
            content_ques = content_ques_text.strip()

            # Process content_ans paragraphs
            paragraphs = content_ans.find_all('p')
            content_ans_text = ''
            for p in paragraphs:
                lines = p.get_text(strip=True).split('\n')
                lines = [line for line in lines if line.strip()]
                if lines:
                    content_ans_text += ' '.join(lines) + ' '
            content_ans = content_ans_text.strip()

            row = {

                'Content_question': content_ques,
                'Answer': content_ans,
            }

            data.append(row)

df = pd.DataFrame(data)
df['Content_question'] = df['Content_question'].astype(str)
df['Answer'] = df['Answer'].astype(str)
df['Content_question'] = df['Content_question'].str.replace(r'<.*?>', '', regex=True)
df['Content_question'] = df['Content_question'].str.replace(r'[\r\n]', '', regex=True)
df['Content_question'] = df['Content_question'].str.replace(r'|Dạ|', '', regex=True)
df['Answer'] = df['Answer'].str.replace(r'<.*?>', '', regex=True)
df['Answer'] = df['Answer'].str.replace(r'[\r\n]', '', regex=True)
df['Answer'] = df['Answer'].str.replace(r'|dạ chào bạn, |, |dạ chào bạn |, |dạ |, |Chào bạn.|, |chào bạn.|, |Chào bạn .|, |chào bạn .|' , '', regex=True)

df.to_csv(filepath+".csv", index=False)
df.to_json(filepath+".json", orient='records', lines=True)

end_time = time.time()  # Stop measuring time
execution_time = end_time - start_time
print("Execution time: {:.2f} seconds".format(execution_time))
```

```python
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

data = []
urls = ["https://benhvienvietduc.org/hoi-dap"]

#path = "/content/drive/MyDrive/1. Academic/1. University/2. Master/4. Bách Khoa - Data Science/1. Courses/Semester 4/Code/Data/Y Tế/"
filename = "data_vietduchospital"
filepath = filename

start_time = time.time()  # Start measuring time

for url in urls:
    response = requests.get(url)
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all elements with class 'item-faq'
    faq_elements = soup.find_all('article', class_='item-faq')

    for faq_element in faq_elements:
        title_element = faq_element.find('div', class_='title-article')
        answer_element = faq_element.find('div', class_='full-content')

        if title_element and answer_element:
            title_text = title_element.get_text(strip=True)
            answer_text = answer_element.get_text(strip=True)

            row = {
                'Content_question': title_text,
                'Answer': answer_text,
            }

            data.append(row)

df = pd.DataFrame(data)
df['Answer'] = df['Answer'].str.replace(r'|Chào bạn, |, |Chào bạn,|, Chào Bạn, |, |Chào Bạn,|,|Chào em, |, |Chào em,|, |Chào anh, |, |Chào anh,|,  |Chào chị, |, |Chào chị,|, ' , '', regex=True)

df.to_csv(filepath + ".csv", index=False)
df.to_json(filepath + ".json", orient='records', lines=True)

end_time = time.time()  # Stop measuring time
execution_time = end_time - start_time
print("Execution time: {:.2f} seconds".format(execution_time))
```

```python
%%time
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Initialize an empty list to store the data
data = []
urls = []
successful_scrapes = 0
failed_scrapes = 0

#path = "/content/drive/MyDrive/1. Academic/1. University/2. Master/4. Bách Khoa - Data Science/1. Courses/Semester 4/Code/Srape data from web/data/"
path = "/content/drive/MyDrive/1. Academic/1. University/2. Master/4. Bách Khoa - Data Science/1. Courses/Semester 4/Code/Srape data from web/data/ivie/"
#filename = "data_ivie_13001to15000"
filepath = path + filename

for i in range(13001, 15000): # 0 upto 5,000 #remove 243 (38931)
    url = 'https://ivie.vn/cong-dong?id=' + str(i)
    urls.append(url)

for url in urls:
    try:
        base_url = url+"&page={}"
        num_pages = 3  # Change this to the number of pages you want to scrape

        for page_num in range(1, num_pages + 1):
            url = base_url.format(page_num)
            response = requests.get(url)
            response.raise_for_status()

            # Check if the request was successful
            if response.status_code == 200:
                html_content = response.text  # Get the HTML content
                soup = BeautifulSoup(html_content, 'html.parser')

                # Rest of your scraping code
                card_bodies = soup.find_all(class_=['card-body'])
                current_sender = None
                current_question = None

                for card_body in card_bodies:
                    sender_element = card_body.find_all(class_=['color-a', 'font-weight-bold', 'sc-eDPEul sc-eldPxv AbVrp jbkqVE'])
                    if sender_element:
                        current_sender = sender_element[0].get_text(strip=True)
                    description_element = card_body.find(class_='description')
                    if description_element:
                        current_question = description_element.get_text(strip=True)
                    data.append([current_sender, current_question])
                    successful_scrapes += 1  # Increment the successful scrape count

    except Exception as e:
        failed_scrapes += 1  # Increment the failed scrape count
        print(f"Error accessing URL: {e}")

# Create a DataFrame from the collected data
df = pd.DataFrame(data, columns=['Sender', 'Question'])
df = df.drop_duplicates()
df = df.reset_index(drop=True)

new_data = {
    "Content_question": [],
    "Answer": []
}

current_col = "Content_question"  # Start with Col1

for index, row in df.iterrows():
    if index == 0:
        new_data["Content_question"].append(row["Question"])
    else:
        prev_sender = df.iloc[index - 1]["Sender"]
        current_sender = row["Sender"]
        if prev_sender == current_sender:
            new_data[current_col][-1] += f" {row['Question']}"  # Join with the previous question
        else:
            current_col = "Content_question" if current_col == "Answer" else "Answer"
            new_data[current_col].append(row["Question"])

# Ensure that both columns have the same length
if len(new_data["Content_question"]) > len(new_data["Answer"]):
    new_data["Answer"].append("")  # Add an empty string to balance the lengths

# Delete the last row if it doesn't have a corresponding question
if new_data["Answer"][-1] == "":
    new_data["Content_question"].pop()
    new_data["Answer"].pop()

# Create the new DataFrame
new_df = pd.DataFrame(new_data)
new_df.to_csv(filepath + ".csv", index=False)
new_df.to_json(filepath+".json", orient='records', lines=True)

# Print the total counts
print(f"Total URLs successfully scraped: {successful_scrapes}")
print(f"Total URLs that could not be accessed: {failed_scrapes}")
```

```python
%%time
path = current_path + "/Thu_Cuc"
# Define the base URL
base_url = 'https://benhvienthucuc.vn/hoi-dap-chuyen-gia/'

# Initialize lists to store the links
get_links = []

# Loop through the pages
for page_number in range(1, 194):  # Adjust the range as needed
    url = f'{base_url}page/{page_number}'

    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        a_tags = soup.find_all('span', class_='span--reply')

        # Extract and append the links
        for a_tag in a_tags:
            link = a_tag.find('a')
            if link:
                item_id = link['href']
                get_links.append(item_id)
    else:
        print(f"Failed to fetch data from {url}")
```

```python
%%time
filename = "data_ThuCuc"
filepath = path + "/" + filename
data = []

# Initialize counters
successful_scrapes = 0
failed_scrapes = 0

for url in get_links:
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an exception for bad responses (e.g., 404)
        html_content = response.text
        soup = BeautifulSoup(html_content, 'html.parser')
        content_question = soup.find('div', class_='entry-cau-hoi__content').get_text(strip=True)
        content_ans_list = soup.find('div', class_='entry-cau-hoi__content entry-content')
        # soup.find_all('p', dir='ltr')

        if content_ans_list:
            content_ans_text = ''
            for p_ans in content_ans_list:
                content_ans_text += p_ans.get_text(strip=True) + '\n'

            row = {
                'Content_question': content_question,
                'Answer': content_ans_text.strip(),
            }
            data.append(row)

            # Increment the successful scrape count and print the URL
            successful_scrapes += 1
            #print(f"Scraped URL: {url}")
    except Exception as e:
        # Increment the failed scrape count and handle the exception
        failed_scrapes += 1
        print(f"Error accessing URL: {url}, {e}")

df = pd.DataFrame(data)
df.to_csv(filepath + ".csv", index=False)
df.to_json(filepath + ".json", orient='records', lines=True)

# Print the total counts
print(f"Total URLs successfully scraped: {successful_scrapes}")
print(f"Total URLs that could not be accessed: {failed_scrapes}")
```

```python
#remove emojis and special characters
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese characters
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
        #u"\u0023-\u0039" # digits and hash
        #u"\u00a9-\u00ff" # latin characters
        #u"\u2000-\u206F" # punctuation and symbols
        u"\u20a0-\u20ff" # currency symbols
        u"\u2100-\u214F" # letter like symbols
        u"\u2190-\u21FF" # arrows
        u"\u2200-\u22FF" # math operators
        u"\u2300-\u23FF" # misc technical
        u"\u25A0-\u25FF" # geometric shapes
        u"\u2600-\u26FF" # misc symbols
        u"\u2700-\u27BF" # dingbats
        u":[\)]|:p+|:D+|:o[\)\(]|:[\)\(]|=\)|=\(|:\'[\(\)]|<3"
        "]+", flags=re.UNICODE)

df_conv_loc['Content_question'] = df_conv_loc['Content_question'].apply(lambda x: emoji_pattern.sub(r'', x))
df_conv_loc['Answer'] = df_conv_loc['Answer'].apply(lambda x: emoji_pattern.sub(r'', x))
```

```python
def preprocess_text(text):

    # Convert text to lowercase
    text = text.lower()

    # Remove sequences of specific alphabet characters
    text = re.sub(r'\b[k]+(?=\s|$)', '', text)
    text = re.sub(r'\b[đ]+(?=\s|$)', '', text)
    text = re.sub(r'\b[h]+(?=\s|$)', '', text)
    text = re.sub(r'\b[dm]+(?=\s|$)', '', text)
    text = re.sub(r'\b[đm]+(?=\s|$)', '', text)
    text = re.sub(r'[\n\t]', '', text)
    # Replace specific text patterns
    text = text.strip()
    text = text.replace(r'cũng z|cũng zậy', 'cũng vậy')
    text = text.replace('đúng r', 'đúng rồi')
    text = text.replace('cx', 'chính xác')
    text = text.replace(r'tks|tx|thanks|cmơn', 'cảm ơn')
    text = text.replace(r"bv", "bệnh viện")
    text = text.replace(r"bác sĩ", "bác sỹ")
    text = text.replace(r"bs", "bác sỹ")
    text = text.replace(r"bhyt", "bảo hiểm y tế")
    text = text.replace(r"bhxh", "bảo hiểm xã hội")
    text = text.replace(r"chào bạn!cảm ơn bạn đã gửi câu hỏi đến icnm", "")
    text = text.replace(r"chào bạn,cảm ơn bạn đã gửi câu hỏi tới icnm,", "")
    text = text.replace(r"chào em,cảm ơn em đã gửi câu hỏi tới icnm,", "")
    text = text.replace(r"chào em!cảm ơn em đã gửi câu hỏi tới icnm,", "")
    text = text.replace(r"cảm ơn anh đã gửi câu hỏi tới icnm,|", "")
    text = text.replace(r"cảm ơn chị đã gửi câu hỏi tới icnm,", "")
    text = text.replace(r"cảm ơn bạn đã gửi câu hỏi tới icnm,", "")
    text = text.replace(r"cảm ơn em đã gửi câu hỏi tới icnm,", "")

    # Remove newline and tab characters
    text = re.sub(r'[\n\t]', '', text)
    return text

df_conv_loc['Content_question'] = df_conv_loc['Content_question'].apply(preprocess_text)
df_conv_loc['Answer'] = df_conv_loc['Answer'].apply(preprocess_text)
```
